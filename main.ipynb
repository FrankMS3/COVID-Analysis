```python
#Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import json
from datetime import datetime
```

```python
#insert the website link 
#hospitalisation, cases number, death and RAT cases.
table_hos = []
table_cases = []
table_death = []
table_rat = []

states = ["vic", "nsw", "qld", "wa", "sa","tas","act", "nt" ]
for x in states:
    url_hos = (f"https://covidlive.com.au/report/daily-hospitalised/{x}")
    url_cases = (f"https://covidlive.com.au/report/daily-cases/{x}")
    url_deaths = (f"https://covidlive.com.au/report/daily-deaths/{x}") 
    url_rat = (f"https://covidlive.com.au/report/daily-cases-probable/{x}") 
    
    # retrieve database:
    table_hos1 = pd.read_html(url_hos)[1]
    table_cases1 = pd.read_html(url_cases)[1]
    table_death1 = pd.read_html(url_deaths)[1]
    table_rat1 = pd.read_html(url_rat)[1]

    #append to the list:
    table_hos.append(table_hos1)
    table_cases.append(table_cases1)
    table_death.append(table_death1)
    table_rat.append(table_rat1)
    
```

```python
#VIC hospitalisation, Cases number, Death number and RAT cases
vic_hos = table_hos[0].copy()
vic_cases = table_cases[0].copy()
vic_death = table_death[0].copy()
vic_rat = table_rat[0].copy()

#NSW hospitalisation, Cases number, Death number and RAT cases
nsw_hos = table_hos[1].copy()
nsw_cases = table_cases[1].copy()
nsw_death = table_death[1].copy()
nsw_rat = table_rat[1].copy()

#QLD hospitalisation, Cases number, Death number and RAT cases
qld_hos = table_hos[2].copy()
qld_cases = table_cases[2].copy()
qld_death = table_death[2].copy()
qld_rat = table_rat[2].copy()

#WA hospitalisation, Cases number, Death number and RAT cases
wa_hos = table_hos[3].copy()
wa_cases = table_cases[3].copy()
wa_death = table_death[3].copy()
wa_rat = table_rat[3].copy()

#SA hospitalisation, Cases number, Death number and RAT cases
sa_hos = table_hos[4].copy()
sa_cases = table_cases[4].copy()
sa_death = table_death[4].copy()
sa_rat = table_rat[4].copy()

#TAS hospitalisation, Cases number, Death number and RAT cases
tas_hos = table_hos[5].copy()
tas_cases = table_cases[5].copy()
tas_death = table_death[5].copy()
tas_rat = table_rat[5].copy()

#ACT hospitalisation, Cases number, Death number and RAT cases
act_hos = table_hos[6].copy()
act_cases = table_cases[6].copy()
act_death = table_death[6].copy()
act_rat = table_rat[6].copy()

#NT hospitalisation, Cases number, Death number and RAT cases
nt_hos = table_hos[7].copy()
nt_cases = table_cases[7].copy()
nt_death = table_death[7].copy()
nt_rat = table_rat[7].copy()
```

```python
#convert Date string to Datetime
vic_hos['DATE'] = pd.to_datetime(vic_hos['DATE'], format ="%d %b %y")
vic_cases['DATE'] = pd.to_datetime(vic_cases['DATE'], format ="%d %b %y")
vic_death['DATE'] = pd.to_datetime(vic_death['DATE'], format ="%d %b %y")
vic_rat['DATE'] = pd.to_datetime(vic_rat['DATE'], format ="%d %b %y")

nsw_hos['DATE'] = pd.to_datetime(nsw_hos['DATE'], format ="%d %b %y")
nsw_death['DATE'] = pd.to_datetime(nsw_death['DATE'], format ="%d %b %y")
nsw_cases['DATE'] = pd.to_datetime(nsw_cases['DATE'], format ="%d %b %y")
nsw_rat['DATE'] = pd.to_datetime(nsw_rat['DATE'], format ="%d %b %y")

qld_hos['DATE'] = pd.to_datetime(qld_hos['DATE'], format ="%d %b %y")
qld_cases['DATE'] = pd.to_datetime(qld_cases['DATE'], format ="%d %b %y")
qld_death['DATE'] = pd.to_datetime(qld_death['DATE'], format ="%d %b %y")
qld_rat['DATE'] = pd.to_datetime(qld_rat['DATE'], format ="%d %b %y")

wa_hos['DATE'] = pd.to_datetime(wa_hos['DATE'], format ="%d %b %y")
wa_cases['DATE'] = pd.to_datetime(wa_cases['DATE'], format ="%d %b %y")
wa_death['DATE'] = pd.to_datetime(wa_death['DATE'], format ="%d %b %y")
wa_rat['DATE'] = pd.to_datetime(wa_rat['DATE'], format ="%d %b %y")

sa_hos['DATE'] = pd.to_datetime(sa_hos['DATE'], format ="%d %b %y")
sa_cases['DATE'] = pd.to_datetime(sa_cases['DATE'], format ="%d %b %y")
sa_death['DATE'] = pd.to_datetime(sa_death['DATE'], format ="%d %b %y")
sa_rat['DATE'] = pd.to_datetime(sa_rat['DATE'], format ="%d %b %y")

tas_hos['DATE'] = pd.to_datetime(tas_hos['DATE'], format ="%d %b %y")
tas_cases['DATE'] = pd.to_datetime(tas_cases['DATE'], format ="%d %b %y")
tas_death['DATE'] = pd.to_datetime(tas_death['DATE'], format ="%d %b %y")
tas_rat['DATE'] = pd.to_datetime(tas_rat['DATE'], format ="%d %b %y")

act_hos['DATE'] = pd.to_datetime(act_hos['DATE'], format ="%d %b %y")
act_death['DATE'] = pd.to_datetime(act_death['DATE'], format ="%d %b %y")
act_cases['DATE'] = pd.to_datetime(act_cases['DATE'], format ="%d %b %y")
act_rat['DATE'] = pd.to_datetime(act_rat['DATE'], format ="%d %b %y")

nt_hos['DATE'] = pd.to_datetime(nt_hos['DATE'], format ="%d %b %y")
nt_cases['DATE'] = pd.to_datetime(nt_cases['DATE'], format ="%d %b %y")
nt_death['DATE'] = pd.to_datetime(nt_death['DATE'], format ="%d %b %y")
nt_rat['DATE'] = pd.to_datetime(nt_rat['DATE'], format ="%d %b %y")
```

```python
#Joining tables to create a single dataframe for each state
#vic
vic_hos_cases = pd.merge (vic_hos, vic_cases, on = "DATE", how = 'right')
vic_death_rat = pd.merge (vic_death, vic_rat, on = "DATE", how ='right')
vic_raw = pd.merge (vic_hos_cases, vic_death_rat, on = "DATE", how = 'right')

#nsw
nsw_hos_cases = pd.merge (nsw_hos, nsw_cases, on = "DATE", how = 'right')
nsw_death_rat = pd.merge (nsw_death, nsw_rat, on = "DATE", how ='right')
nsw_raw = pd.merge (nsw_hos_cases, nsw_death_rat, on = "DATE", how = 'right')

#QLD
qld_hos_cases = pd.merge (qld_hos, qld_cases, on = "DATE", how = 'right')
qld_death_rat = pd.merge (qld_death, qld_rat, on = "DATE", how ='right')
qld_raw = pd.merge (qld_hos_cases, qld_death_rat, on = "DATE", how = 'right')

#WA
wa_hos_cases = pd.merge (wa_hos, wa_cases, on = "DATE", how = 'right')
wa_death_rat = pd.merge (wa_death, wa_rat, on = "DATE", how ='right')
wa_raw = pd.merge (wa_hos_cases, wa_death_rat, on = "DATE", how = 'right')

#SA
sa_hos_cases = pd.merge (sa_hos, sa_cases, on = "DATE", how = 'right')
sa_death_rat = pd.merge (sa_death, sa_rat, on = "DATE", how ='right')
sa_raw = pd.merge (sa_hos_cases, sa_death_rat, on = "DATE", how = 'right')

#TAS
tas_hos_cases = pd.merge (tas_hos, tas_cases, on = "DATE", how = 'right')
tas_death_rat = pd.merge (tas_death, tas_rat, on = "DATE", how ='right')
tas_raw = pd.merge (tas_hos_cases, tas_death_rat, on = "DATE", how = 'right')

#ACT
act_hos_cases = pd.merge (act_hos, act_cases, on = "DATE", how = 'right')
act_death_rat = pd.merge (act_death, act_rat, on = "DATE", how ='right')
act_raw = pd.merge (act_hos_cases, act_death_rat, on = "DATE", how = 'right')

#NT
nt_hos_cases = pd.merge (nt_hos, nt_cases, on = "DATE", how = 'right')
nt_death_rat = pd.merge (nt_death, nt_rat, on = "DATE", how ='right')
nt_raw = pd.merge (nt_hos_cases, nt_death_rat, on = "DATE", how = 'right')
```

```python
#rename & reorder columns name
vic_rename = vic_raw.rename(columns={"NEW_x": "NEW CASES", "CASES_x": "CURRENT CASES", "NEW_y": "NEW RAT CASES", "CASES_y": "CURRENT RAT CASES"})
vic_reorder = vic_rename[["DATE", "HOSP", "ICU", "VENT", "NEW CASES", "CURRENT CASES", "DEATHS", "NEW RAT CASES", "CURRENT RAT CASES"]]
#convert "New Rat Cases" column from object to float
vic_reorder["NEW RAT CASES"]= pd.to_numeric(vic_reorder["NEW RAT CASES"], errors='coerce')

#NSW
nsw_rename = nsw_raw.rename(columns={"NEW_x": "NEW CASES", "CASES_x": "CURRENT CASES", "NEW_y": "NEW RAT CASES", "CASES_y": "CURRENT RAT CASES"})
nsw_reorder = nsw_rename[["DATE", "HOSP", "ICU", "VENT", "NEW CASES", "CURRENT CASES", "DEATHS", "NEW RAT CASES", "CURRENT RAT CASES"]]
#convert "New Rat Cases" column from object to float
nsw_reorder["NEW RAT CASES"]= pd.to_numeric(nsw_reorder["NEW RAT CASES"], errors='coerce')

#QLD
qld_rename = qld_raw.rename(columns={"NEW_x": "NEW CASES", "CASES_x": "CURRENT CASES", "NEW_y": "NEW RAT CASES", "CASES_y": "CURRENT RAT CASES"})
qld_reorder = qld_rename[["DATE", "HOSP", "ICU", "VENT", "NEW CASES", "CURRENT CASES", "DEATHS", "NEW RAT CASES", "CURRENT RAT CASES"]]
#convert "New Rat Cases" column from object to float
qld_reorder["NEW RAT CASES"]= pd.to_numeric(qld_reorder["NEW RAT CASES"], errors='coerce')

#WA
wa_rename = wa_raw.rename(columns={"NEW_x": "NEW CASES", "CASES_x": "CURRENT CASES", "NEW_y": "NEW RAT CASES", "CASES_y": "CURRENT RAT CASES"})
wa_reorder = wa_rename[["DATE", "HOSP", "ICU", "VENT", "NEW CASES", "CURRENT CASES", "DEATHS", "NEW RAT CASES", "CURRENT RAT CASES"]]
#convert "New Rat Cases" column from object to float
wa_reorder["NEW RAT CASES"]= pd.to_numeric(wa_reorder["NEW RAT CASES"], errors='coerce')

#SA
sa_rename = sa_raw.rename(columns={"NEW_x": "NEW CASES", "CASES_x": "CURRENT CASES", "NEW_y": "NEW RAT CASES", "CASES_y": "CURRENT RAT CASES"})
sa_reorder = sa_rename[["DATE", "HOSP", "ICU", "VENT", "NEW CASES", "CURRENT CASES", "DEATHS", "NEW RAT CASES", "CURRENT RAT CASES"]]
#convert "New Rat Cases" column from object to float
sa_reorder["NEW RAT CASES"]= pd.to_numeric(sa_reorder["NEW RAT CASES"], errors='coerce')

#TAS
tas_rename = tas_raw.rename(columns={"NEW_x": "NEW CASES", "CASES_x": "CURRENT CASES", "NEW_y": "NEW RAT CASES", "CASES_y": "CURRENT RAT CASES"})
tas_reorder = tas_rename[["DATE", "HOSP", "ICU", "VENT", "NEW CASES", "CURRENT CASES", "DEATHS", "NEW RAT CASES", "CURRENT RAT CASES"]]
#convert "New Rat Cases" column from object to float
tas_reorder["NEW RAT CASES"]= pd.to_numeric(tas_reorder["NEW RAT CASES"], errors='coerce')

#ACT
act_rename = act_raw.rename(columns={"NEW_x": "NEW CASES", "CASES_x": "CURRENT CASES", "NEW_y": "NEW RAT CASES", "CASES_y": "CURRENT RAT CASES"})
act_reorder = act_rename[["DATE", "HOSP", "ICU", "VENT", "NEW CASES", "CURRENT CASES", "DEATHS", "NEW RAT CASES", "CURRENT RAT CASES"]]
#convert "New Rat Cases" column from object to float
act_reorder["NEW RAT CASES"]= pd.to_numeric(act_reorder["NEW RAT CASES"], errors='coerce')

#NT
nt_rename = nt_raw.rename(columns={"NEW_x": "NEW CASES", "CASES_x": "CURRENT CASES", "NEW_y": "NEW RAT CASES", "CASES_y": "CURRENT RAT CASES"})
nt_reorder = nt_rename[["DATE", "HOSP", "ICU", "VENT", "NEW CASES", "CURRENT CASES", "DEATHS", "NEW RAT CASES", "CURRENT RAT CASES"]]
#convert "New Rat Cases" column from object to float
nt_reorder["NEW RAT CASES"]= pd.to_numeric(nt_reorder["NEW RAT CASES"], errors='coerce')
```
